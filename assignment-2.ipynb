# Scraping top 5 Wikipedia movie website

```python
import csv
import pandas as pd
import requests
import bs4
```

```python
res = requests.get("https://en.wikipedia.org/wiki/Tenet_(film)")
res1 = requests.get("https://en.wikipedia.org/wiki/Bad_Boys_for_Life")
res2 = requests.get("https://en.wikipedia.org/wiki/Dolittle_(film)")
res3 = requests.get("https://en.wikipedia.org/wiki/A_Little_Red_Flower")
res4 = requests.get("https://en.wikipedia.org/wiki/Godzilla_vs._Kong")
```

```python
soup = bs4.BeautifulSoup(res.text,"lxml")
soup1 = bs4.BeautifulSoup(res1.text,"lxml")
soup2 = bs4.BeautifulSoup(res2.text,"lxml")
soup3 = bs4.BeautifulSoup(res3.text,"lxml")
soup4 = bs4.BeautifulSoup(res4.text,"lxml")
```

```python
print("1'ST MOVIE\n")
#title
title1 = soup.select("title")[0].getText()
print("Movie name: ",title1[0:5])
#Director
for item1 in soup.select(".infobox-data")[0]:
    print("Director name: ",item1.text)
#Year of release
for year1 in soup.select(".infobox-data")[9]:
    print("Year of release:",year1.text[12:16])
#Overall gross rating
gross1 =soup.select(".infobox-data")[14].get_text()
print("Overall gross rating:",gross1[0:14])
#The cast
cast1 = soup.select(".infobox-data")[3].get_text()
print("Cast:",cast1.split("\n"))

print("\n")
print("2'ND MOVIE\n")

#title
title2 = soup1.select("title")[0].getText()
print(title2[0:18])
#Director
for item2 in soup1.select(".infobox-data")[0]:
    print("Director name: ",item2.text)
#Year of release
for year2 in soup1.select(".infobox-data")[11]:
    print("Year of release:",year2.text[12:17])
#Overall gross rating
gross2 = soup1.select(".infobox-data")[16].get_text()
print("Overall gross rating:",gross2[0:14])
#The cast
cast2 = soup1.select(".infobox-data")[5].get_text()
print("Cast:",cast2.split("\n"))
print("\n")

print("3'RD MOVIE\n")

#title
title3 = soup2.select("title")[0].getText()
print("Movie name: ",title3[0:5])
#Director
for item3 in soup2.select(".infobox-data")[0]:
    print("Director name: ",item3.text)
#Year of release
for year3 in soup2.select(".infobox-data")[11]:
    print("Year of release:",year3.text[12:17])
#Overall gross rating
gross3 =soup2.select(".infobox-data")[16].get_text()
print("Overall gross rating:",gross3[0:14])
#The cast
cast3 = soup2.select(".infobox-data")[5].get_text()
print("Cast:",cast3.split("\n"))
print("\n")

print("4'TH MOVIE\n")
#title
title4 = soup3.select("title")[0].getText()
print("Movie name: ",title4[0:10])
#Director
for item4 in soup3.select(".infobox-data")[0]:
    print("Director name: ",item4.text)
#Year of release
for year4 in soup3.select(".infobox-data")[11]:
    print("Year of release:",year4.text[13:19])
#Overall gross rating
gross4 =soup3.select(".infobox-data")[-1].get_text()
print("Overall gross rating:",gross4[0:12])
#The cast
cast4 = soup3.select(".infobox-data")[7].get_text()
print("Cast:",cast4.split("\n"))
print("\n")

print("5'TH MOVIE\n")

#title
title5 = soup4.select("title")[0].getText()
print("Movie name: ",title5[0:18])
#Director
for item5 in soup4.select(".infobox-data")[0]:
    print("Director name: ",item5.text)
#Year of release
for year5 in soup4.select(".infobox-data")[11]:
    print("Year of release:",year5.text[10:16])
#Overall gross rating
gross5 =soup4.select(".infobox-data")[17].get_text()
print("Overall gross rating:",gross5[10:14])
#The cast
cast5 = soup4.select(".infobox-data")[5].get_text()
print("Cast:",cast5.split("\n"))
```
